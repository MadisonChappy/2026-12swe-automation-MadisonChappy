## The scenario

A tutoring company wants to predict a student's exam score based on how many hours they studied.

They have collected data from previous students. We are going to use that data to train an ML model, then use the model to predict scores for new students.

## Step 1 — Load the tools

We need two libraries:
- **NumPy** — for working with arrays of numbers
- **scikit-learn** — a machine learning library. We are importing the `LinearRegression` class from it.

````python
import numpy as np
from sklearn.linear_model import LinearRegression

print("Libraries loaded successfully.")
````

## Step 2 — Load the training data

Here is the data from previous students:

| Hours studied | Exam score |
|:---:|:---:|
| 1 | 45 |
| 2 | 52 |
| 3 | 58 |
| 4 | 65 |
| 5 | 70 |
| 6 | 78 |
| 7 | 84 |
| 8 | 91 |

- `X` = the **feature** (input) — hours studied
- `y` = the **target** (output) — exam score

> Notice that `X` is a 2D array (a list of lists). scikit-learn requires this format for features.

````python
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])  # hours studied
y = np.array([45, 52, 58, 65, 70, 78, 84, 91])           # exam scores

print(f"Training examples: {len(y)}")
print(f"Features (X): {X.flatten()}")
print(f"Targets  (y): {y}")
````

## Step 3 — Create the model

We create a `LinearRegression` object. This is an ML model. At this point it does not know anything — it hasn't seen any data yet.

> This is **object-oriented programming** in action. `LinearRegression` is a class. `model` is an instance (object) of that class.

````python
model = LinearRegression()

print("Model created:", model)
````

## Step 4 — Train the model

We call `model.fit(X, y)` to train the model on our data.

The model analyses the relationship between hours studied and exam scores, and finds the best mathematical line to describe that relationship.

````python
model.fit(X, y)

print("Training complete.")
print(f"Slope (coefficient): {model.coef_[0]:.2f}")
print(f"Intercept:           {model.intercept_:.2f}")
````

## Step 5 — Make a prediction

Now we use `model.predict()` to predict the score for a student who studied for 10 hours.

> The model has **never seen** a student who studied for 10 hours. It is using the pattern it learned to make an educated guess.

````python
hours = np.array([[10]])
predicted_score = model.predict(hours)

print(f"Predicted score for 10 hours of study: {predicted_score[0]:.1f}")
````

## Step 6 — Visualise the model

Let's plot the training data and the model's line of best fit.

````python
import matplotlib.pyplot as plt

X_line = np.linspace(0, 12, 100).reshape(-1, 1)
y_line = model.predict(X_line)

plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='steelblue', zorder=5, label='Training data')
plt.plot(X_line, y_line, color='tomato', label='Model prediction line')
plt.scatter([10], predicted_score, color='green', marker='D', s=100,
            zorder=6, label=f'Prediction at 10 hrs: {predicted_score[0]:.1f}')
plt.xlabel('Hours studied')
plt.ylabel('Exam score')
plt.title('Hours Studied vs Predicted Exam Score')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````
